```python
# 2.Create a histogram for mean radius. What does the distribution look like? Is it skewed? If so, in which direction?
plt.figure(figsize=(10,6))
plt.hist(df_breast_cancer_data['mean radius'])
plt.title("Histogram for mean Radius")
plt.xlabel("Mean Radius")
plt.ylabel("Frequency")
plt.show()

# Observation
# The hist plot is right skewed indicating that there are many small radius tumors and few tumors having
# large radius.
```

```python
# 3.Use a boxplot to visualize the distribution of mean texture. Are there any outliers in the data? What is the interquartile range (IQR)?

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.boxplot(x= df_breast_cancer_data['mean texture'])
plt.title("Boxplot for mean texture")
plt.xlabel("Mean Texture")
plt.ylabel("Frequency")

plt.subplot(1,3,2)
plt.boxplot(x= df_breast_cancer_data['mean area'])
plt.title("Boxplot for mean area")
plt.xlabel("Mean Area")
plt.ylabel("Frequency")


plt.subplot(1,3,3)
plt.boxplot(x= df_breast_cancer_data['mean compactness'])
plt.title("Boxplot for mean compactness ")
plt.xlabel("Mean Compactness")
plt.ylabel("Frequency")
plt.show()

# All above features indicates presnce of outliers i.e there are few data sets with large tumors having large area texture and compactness.
```

```python
# Finding Inter quartile range
Q1 = df_breast_cancer_data['mean texture'].quantile(0.25)
Q3 =  df_breast_cancer_data['mean texture'].quantile(0.75)

Iqr = Q3 - Q1

Lower_bound = Q1 - 1.5 * Iqr
Upper_bound = Q3 + 1.5 * Iqr

print(f"Inter Quartile range : {Iqr:.2f}")
print(f"Lower Bound : {Lower_bound:.2f}")
print(f"Upper Bound : {Upper_bound:.2f}")
```

```python
# 4.Group the data by the target variable and calculate the mean for mean perimeter. Is there a noticeable difference between the mean perimeter of benign and malignant cases?

df_breast_cancer_data.groupby('target')['mean perimeter'].mean()

# Observation:
# There is noticeable difference between the mean perimeter of benign and malignant case.
```

5. Feature Engineering

1.Create a new feature called `area_to_perimeter_ratio` by dividing `mean area` by `mean perimeter`. How does this new feature differ between benign and malignant cases?

2.Analyze the impact of `area_to_perimeter_ratio` on the classification of cancer. Does this new feature provide any additional insights or separability between benign and malignant cases?

3.Are there other combinations of features that could be useful? Create at least one more derived feature and analyze its impact on the classification

```python
# Feature Engineering
df_breast_cancer_data['area_to_perimeter_ratio'] = (df_breast_cancer_data['mean area'] / df_breast_cancer_data['mean perimeter'])

df_breast_cancer_data.groupby('target')['area_to_perimeter_ratio'].mean()

# Observations
# area_to_perimeter_ratio clearly indicates more is the ratio chances of malignant tumor is more.
```

```python
# Another derived feature
df_breast_cancer_data['radius_texture_interaction'] = (
    df_breast_cancer_data['mean radius'] * df_breast_cancer_data['mean texture']
)

df_breast_cancer_data.groupby('target')['radius_texture_interaction'].mean()


```

```python
# new feature worst radius_texture_interaction can be helpful for classification as well
df_breast_cancer_data['worst radius_texture_interaction'] = (
    df_breast_cancer_data['worst radius'] * df_breast_cancer_data['worst texture']
)

df_breast_cancer_data.groupby('target')['worst radius_texture_interaction'].mean()

# worst radius and worst texture shows more distinct results.
```

### for else is used for search item cases along with break statement in for search if values not found and need to give ay message or perform any action in that case for else is used.


```python
def search_item(item_list,target_item):
  for item in item_list:
    if item == target_item:
      print("item found")
      break
  else:
    print("item not found")


search_item([1,2,3,4,5,6],4)

search_item([1,2,3,4,5,6],10)
```

```python
df_hour3 = df_hour2.copy()
df_hour3.mnth = df_hour3.mnth.replace([5,6,7,8,9],5)
np.unique(df_hour3.mnth)
```

```python
df_hour3.hr = df_hour3.hr.replace([0,1,2,3,4,5],5)
df_hour3.hr = df_hour3.hr.replace([11,12,13,14,15],11)

np.unique(df_hour3.hr)
```

```python


cat_cols = ['season', 'weathersit', 'weekday', 'mnth', 'hr']
df_hour3 = pd.get_dummies(df_hour3, columns=cat_cols, drop_first=True)
df_hour3
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df_hour3, train_size = 0.7, random_state = 100)
```

```python
y_train = df_train.pop("cnt") # pop will return column to new datframe and delete from column  from parent dataframe
X_train = df_train

y_test = df_test.pop("cnt")
X_test = df_test
```

```python
y_train.info()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_train_pred= lr.predict(X_train)
r2_score(y_train, y_train_pred)
```

```python
y_test_pred= lr.predict(X_test)
r2_score(y_test, y_test_pred)
```
